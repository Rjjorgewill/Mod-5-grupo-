#Lista de criptomoedas selecionadas para fazermos a análise no período de 2015 a 2018: <br><br>
1. **Bitcoin (BTC)**: Lançada em 2009, o Bitcoin foi a primeira criptomoeda e continua sendo a mais conhecida e valiosa até hoje. <br>

2. **Ethereum (ETH)**: Lançada em 2015, o Ethereum é uma plataforma blockchain que permite a criação e execução de contratos inteligentes e aplicativos descentralizados. <br>

3. **Ripple (XRP)**: Criada em 2012, a Ripple é uma rede de pagamento em tempo real que utiliza a criptomoeda XRP para facilitar transferências internacionais de dinheiro. <br>

4. **Litecoin (LTC)**: Lançada em 2011, o Litecoin foi uma das primeiras altcoins a surgir após o Bitcoin. Ela se diferencia por ter tempos de bloqueio mais rápidos e um algoritmo de mineração diferente. <br>

5. **Dash (DASH)**: Inicialmente lançada como Darkcoin em 2014 e posteriormente renomeada para Dash (Digital Cash), é uma criptomoeda que oferece maior privacidade e transações instantâneas. <br>

6. **Monero (XMR)**: Lançada em 2014, a Monero é conhecida por seu foco em privacidade e anonimato, usando tecnologias como assinaturas de anel e endereços ocultos. <br>

7. **NEM (XEM)**: Lançada em 2015, a NEM é uma plataforma blockchain que oferece recursos avançados, como criação de ativos digitais personalizados e mensagens criptografadas. <br>

8. **Stellar (XLM)**: Criada em 2014, a Stellar é uma plataforma de pagamento e remessa que visa facilitar transferências de dinheiro de forma rápida, barata e segura. <br>

9. **Cardano (ADA)**: Lançada em 2017, a Cardano é uma plataforma blockchain que busca oferecer maior segurança, escalabilidade e sustentabilidade para a execução de contratos inteligentes. <br>

10. **EOS (EOS)**: Lançada em 2018, a EOS é uma plataforma blockchain que visa oferecer escalabilidade e facilidade de uso para o desenvolvimento de aplicativos descentralizados. <br><br><br>

````python
#renomeando a coluna 'Market Cap' no dataset dash:
dash.rename(columns={'Market Cap': 'Marketcap'}, inplace=True)
dash
````

````python
#criando as colunas 'Name' e 'Symbol':

name = 'Dash'
simbolo = 'DASH'
dash = dash.assign(Name=name)
dash = dash.assign(Symbol=simbolo)
dash
````

````python
#como vimos que os dados estão invertidos (os ultimos em primeiro, no formato de pilha), vamos mudar a ordem:
dash = dash.iloc[::-1]
dash
````

````python
#Criar a coluna SNo:
dash['SNo'] = dash.reset_index().index
dash
````

````python
#para colocar na mesma ordem:
ordem = ['SNo', 'Name', 'Symbol', 'Date', 'High', 'Low', 'Open', 'Close', 'Volume', 'Marketcap']

dash = dash[ordem]
dash
````

````python
#colocando a coluna 'Date' no mesmo padrão das outras:
dash['Date'] = pd.to_datetime(dash['Date'])

# Formatar a coluna 'Date' no formato desejado
dash['Date'] = dash['Date'].dt.strftime('%Y-%m-%d 23:59:59')
dash
````

````python
#colocando as colunas 'High', 'Low', 'Open' e 'Close' do dataframe dash para o
#mesmo formato dos outros dataframes.
#para que valores como -> 1,001.50 se torne 1001.50 e não cause erros
dash['High'] = dash['High'].str.replace(',', '').str.replace('.', '.')
dash['Low'] = dash['Low'].str.replace(',', '').str.replace('.', '.')
dash['Open'] = dash['Open'].str.replace(',', '').str.replace('.', '.')
dash['Close'] = dash['Close'].str.replace(',', '').str.replace('.', '.')

dash['High'] = dash['High'].astype(float)
dash['Low'] = dash['Low'].astype(float)
dash['Open'] = dash['Open'].astype(float)
dash['Close'] = dash['Close'].astype(float)
dash
````

####Vamos fazer a análise da autocorrelação de cada uma das colunas numéricas do dataframe para identificar-mos padrões sazonais, tendências e padrões de repetição ao longo do tempo, caso haja. <br><br>
Autocorrelação dos dados:

````python
#transformando as colunas em numericas
criptos_final['Alta'] = criptos_final['Alta'].str.replace(',', '.').astype(float)
criptos_final['Baixa'] = criptos_final['Baixa'].str.replace(',', '.').astype(float)
criptos_final['Fechado'] = criptos_final['Fechado'].str.replace(',', '.').astype(float)
criptos_final['Aberto'] = criptos_final['Aberto'].str.replace(',', '.').astype(float)

#agrupando as moedas
grupos = criptos_final.groupby('Nome')

autocorr = {}
for criptomoeda, grupo in grupos:
    autocorr[criptomoeda] = grupo.corr()

for criptomoeda, corr in autocorr.items():
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

    plt.ylim(len(corr), 0)
    plt.xlim(0, len(corr))
    plt.title(f"{criptomoeda}")
    plt.show()
````
